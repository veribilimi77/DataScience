### **Hipotez nasıl oluşturulur?**

#### **Mağaza Seviyesi Hipotezleri**
1. **Şehir tipi**: Kentsel veya kırsal şehirlerde bulunan mağazalar, oradaki insanların yüksek gelir seviyesinden dolayı daha yüksek satışlara sahip olmalıdır.

2. **Nüfus Yoğunluğu**: Yoğun nüfuslu bölgelerde bulunan mağazaların daha fazla talep nedeniyle daha fazla satış yapması gerekiyor. 
3. **Mağaza Kapasitesi**: Çok büyük mağazalar,daha yüksek satışlara sahip olmalı ve insanlar her şeyi tek bir yerden almayı tercih edeceklerdir.

4. **Rakipler**: Yakındaki benzer mağazalara sahip olan mağazalar, daha fazla rekabet nedeniyle daha az satış yapması beklenir.

5. **Pazarlama**: Pazarlama bölümü iyi olan mağazalar, müşterileri doğru teklifler ve reklamlar aracılığıyla çekebileceklerinden daha yüksek satışlara sahip olmalıdır.

6. **Yer**: Popüler pazarlarda yer alan mağazalar, müşterilere daha iyi erişim nedeniyle daha yüksek satışlara sahip olmalıdır.

7. **Ambiyans**: Kibar ve mütevazi insanlar tarafından bakımlı ve yönetilen mağazaların daha fazla yer kaplaması ve dolayısıyla daha yüksek satış yapması bekleniyor.

### **Ürün Seviyesi Hipotezleri**
1. **Marka**: Markalı ürünlerin müşteride olan güveninin yüksek olması nedeniyle satışları yüksek olmalıdır.

2. **Ambalaj**: Ambalajı iyi olan ürünler müşterileri çekebilir ve daha fazlasını satabilir.

3. **Faydası**: Günlük kullanım ürünlerinin, özel kullanım ürünlerine göre daha yüksek satış eğilimi göstermesi gerekir.

4. **Sergileme Alanı**: Mağazada daha büyük raflar verilen ürünlerin önce dikkat çekmesi ve daha fazla satış yapması muhtemeldir.
5. **Mağazada Görünürlük**: Bir mağazada ürünün konumu, satışları etkileyecektir. Girişte olanlar arkadakilere göre müşterinin dikkatini daha fazla çekecektir.

6. **Reklam**: Mağazada ürünlerin daha iyi reklamı, çoğu durumda daha yüksek satışlar yapmalıdır.
7. **Promosyon Teklifleri**: Cazip tekliflerle birlikte sunulan ürünler ve indirimler daha fazlasını satacaktır.

### **Müşteri Seviyesi Hipotezleri**
1. **Müşteri Davranışı**: Müşterilerin yerel ihtiyaçlarını karşılamak için doğru ürün grubunu bulunduran mağazalar daha yüksek satışlara sahip olacaktır.

2. **İş Profili**: Yönetici düzeyinde çalışan müşteri, giriş düzeyinde veya orta düzeyde çalışan müşterilere kıyasla daha yüksek miktarda ürün satın alma şansına sahip olacaktır.

3. **Aile Büyüklüğü**: Daha fazla aile üyesi sayısı, bir müşteri tarafından ürün satın almak için daha fazla harcama yapacaktır.

4. **Yıllık Gelir**: Bir müşterinin yıllık geliri arttıkça, müşterinin yüksek maliyetli ürünler alması daha olasıdır. 
5. **Satın Alma Geçmişi**: Bu bilgilerin mevcut olması, bir kullanıcı tarafından satın alınan bir ürünün sıklığını belirlememize yardımcı olabilir.

### **Makro Seviye Hipotezleri**
**Çevre**: Çevre devlet tarafından güvenli olarak ilan edilirse, müşterinin çevresel problem olmadığından endişe duymayacağından dolayı ürün satın alma olasılığı daha yüksek olacaktır.

**Ekonomik Büyüme**: Eğer mevcut ekonomi tutarlı bir büyüme gösteriyorsa, kişi başına düşen gelir artacaktır, bu nedenle müşterilerin satın alma gücü artacaktır.

    - 3. **Öngörü Modelini Oluşturma:** Bir değişkende eksik verileri doldurmak için bir tahmin modeli bile yapabiliriz. Burada eksik verileri olan değişkeni hedef değişken olarak ve diğer değişkenleri yordayıcı(tahmin edici) olarak ele alacağız. Verilerimizi 2 veri setine ayıracağız - biri bu değişken için eksik değeri olmayan, diğerivise bu değişken için eksik değerleri olan bir veri seti. Eski set, öngörücü modeli oluşturmak için eğitim seti olarak kullanılacak ve daha sonra eksik değerleri tahmin etmek için son  oluşturduğumu veri setine uygulanacaktır.

**Kategorik değişkenleri kodlamak neden önemlidir?

Makine öğrenme algoritmalarının çoğu, sayısal değişkenlerle daha iyi sonuç verir. Bu nedenle, verilerde mevcut kategorik değişkenleri sayısal değişkene dönüştürmek önemlidir. Yapılabilecek bir şey, kategorik değişkenleri tamamen kaldırmaktır, ancak bu, çok fazla bilgi kaybına yol açacaktır. Neyse ki kategorik değişkenlerle başa çıkabilmek için daha akıllı teknikler var.

- **Etiket kodlaması**, Her kategorik değişkeni sayıya dönüştürmek gelir. Sıralı değişkenler için daha uygundur - bazı düzende kategorik değişkenler içinde kullanılabilir.

- **Bir sıcak kodlamada**(One-Hot-encoding); Bir kategorik değişkenin her kategorisi yeni bir ikili sütuna (1/0) dönüştürülür.

**Kategorik değişkenler için etiket kodlaması**

```r
# Etiketleme yaptıktan sonra kategorik değişkenleri siliyoruz

combi[, c("Outlet_Size", "Outlet_Location_Type") := NULL]
```

```r
#Kategorik değişkenleri (One hot encoding) olarak kodluyoruz
ohe = dummyVars("~.", data = combi[,-c("Item_Identifier", "Outlet_Establishment_Year", "Item_Type")], fullRank = T) 
ohe_df = data.table(predict(ohe, combi[,-c("Item_Identifier", "Outlet_Establishment_Year", "Item_Type")])) 
combi = cbind(combi[,"Item_Identifier"], ohe_df)
```

### **Doğrusal regresyon**, 
Yordayıcı(tahminci) modelleme için en basit ve en yaygın kullanılan istatistiksel tekniktir. Aşağıda verilen doğrusal regresyon denklemi:

Y = Q1.X1 + Q(theta)2.X2...Qn.Xn

X1, X2,…, Xn bağımsız değişkenler, Y ise hedef değişken ve tüm tetalar ise katsayılardır. Bir katsayının diğer katsayılara göre büyüklüğü o sayıya karşılık gelen bağımsız değişkenin önemini belirler.

İyi bir doğrusal regresyon modeli için, veriler [bu makalede](https://www.analyticsvidhya.com/blog/2016/07/deeper-regression-analysis-assumptions-plots-solutions/) ayrıntılı olarak bulunabilecek birkaç varsayımı karşılamalıdır. Bu varsayımlardan biri, çok kutupluluğun olmayışıdır, yani bağımsız değişkenler ilişkilendirilmelidir. Bununla birlikte, yukarıdaki korelasyon grafiğine göre, verilerimizde birbiriyle oldukça ilişkili birkaç bağımsız değişken var. Çok kutupluluk konusu düzenlileştirme ile ele alınabilir.

Şimdilik doğrusal regresyon modelimizi tüm değişkenlerle birlikte oluşturalım. Yapacağımız tüm modellerde 5-fold (kat) çapraz doğrulama kullanacağız. Temel olarak çapraz onaylama, bir modelin görünmeyen verileri ne kadar iyi genellediği konusunda bir fikir verir.

#### k-kat Çapraz Doğrulama
Sonraki bölümlerde, modellerimizi test etmek için k-kat çapraz doğrulama (CV) kullanacağız. K-katlama CV için izlenen adımlar aşağıdaki gibidir:

 1 - Verileri rastgele k ”kıvrımlara” bölün.
 2 - Veri kümenizdeki her bir k-kat için modelinizi veri kümesinin k - 1 katları üzerine yerleştirin. Ardından, birinci katın etkinliğini kontrol etmek için modeli test edin.
 3 - Her bir tahminde gördüğünüz hatayı kaydedin.
 4 - K-kıvrımlarının her biri test seti olarak görev yapana kadar bunu tekrarlayın.
 K kaydedilen hatalarınızın ortalaması çapraz doğrulama hatası olarak adlandırılır ve model için performans ölçütünüz olarak kullanılır.

Aşağıda, k = 10 olduğunda bir k-katlama doğrulamasının görselleştirilmesi gösterilmektedir.

### **RandomForest,** 
Güçlü bir tahmin modeli oluşturmak için belirli sayıda zayıf öğrencinin (karar ağaçları) birleştirildiği ağaç tabanlı bir önyükleme algoritmasıdır. Her bir bireysel öğrenci için, bir karar ağacı modeli oluşturmak için rastgele bir sıra örneği ve rastgele seçilmiş birkaç değişken kullanılır. Son tahmin, bireysel öğreniciler tarafından yapılan tüm tahminlerin bir fonksiyonu olabilir. Bir regresyon probleminde, nihai tahmin, tüm tahminlerin ortalaması olabilir. Ayrıntılı açıklama için bu makaleyi ziyaret edin.

Şimdi 400 ağaçlı bir RandomForest modeli inşa edeceğiz. Burada kullanılan diğer ayar parametreleri mtry - no. prediktör değişkenlerinin rastgele her bölünmede örneklenmesi ve min.node.size - minimum terminal düğüm büyüklüğü (bu sayıyı büyük ayarlamak daha küçük ağaçlara neden olur ve fazla sığmayı azaltır).

```r
set.seed(1237) 
my_control = trainControl(method="cv", number=5) # 5-fold CV 
tgrid = expand.grid(
  .mtry = c(3:10),
  .splitrule = "variance",
  .min.node.size = c(10,15,20)
 )
rf_mod = train(x = train[, -c("Item_Identifier", "Item_Outlet_Sales")],
                y = train$Item_Outlet_Sales,
               method='ranger',
                trControl= my_control,
                tuneGrid = tgrid,
               num.trees = 400,
               importance = "permutation")
```

Lider panosundaki puanımız RandomForest kullanarak oldukça gelişti. Şimdi farklı ayarlama parametreleri için RMSE puanlarını görelim.

```r
#En iyi Model Parametereleri
plot(rf_mod)
```

Yukarıda gösterilen grafiğe göre, en iyi puan mtry = 5 ve min.node.size = 20'de elde edilmiştir.

**Değişken Önemi**
RandomForest modeline dayanarak grafikte değişkenlerin önem bilgilerini görselleştirelim.

**XGBoost** hızlı ve verimli bir algoritmadır ve birçok veri bilimi yarışmasının galipleri tarafından kullanılmıştır. Bu bir artırma algoritmasıdır ve artırma hakkında daha fazla bilgi için [bu makaleye](https://www.analyticsvidhya.com/blog/2015/11/quick-introduction-boosting-algorithms-machine-learning/) başvurabilirsiniz. XGBoost yalnızca sayısal değişkenlerle çalışır bu sebeple kategorik değişkenleri sayısal değişkenlerle değiştirdik. XGBoost'ta Genel Parametreler, Güçlendirici Parametreler ve Görev Parametreleri altında genel olarak sınıflandırılabilecek birçok ayar parametresi vardır.

- **Genel parametreler**, yükseltme yapmak için hangi hidroforu kullandığımızı gösterir. Yaygın olarak kullanılan ağaç veya doğrusal modeldir
- **Yükseltici parametreleri**, seçtiğiniz yükselticiye göre değişir Öğrenme senaryosuna karar veren - **Öğrenme Görevi parametreleri**, örneğin, regresyon görevleri, sıralama görevlerinde farklı parametreler kullanabilir.

**Modelimizde kullanacağımız parametrelere bir göz atalım.

**eta:** Öğrenme hızı veya büzülme faktörü olarak da bilinir. Aslında, yükseltme işlemini daha tutucu hale getirmek için özellik ağırlıklarını küçültür. Aralık, 0 ile 1 arasındadır. Düşük eta değeri, modelin aşırı yüklenmeye daha dayanıklı olduğu anlamına gelir.
**gamma:** Aralık, 0 ile ∞ arasındadır. Gama büyüdükçe algoritma daha tutucu olur.
**max_depth:** Bu parametreyi kullanarak bir ağacın maksimum derinliğini belirleyebiliriz.
**alt örnek:** Modelin rasgele ağaçları büyütmek için seçeceği satırların oranıdır.
**colsample_bytree:** Modeldeki her bir ağacı oluşturmak için rastgele seçilen değişkenlerin oranıdır.

```r
param_list = list(objective = "reg:linear", eta=0.01, gamma = 1,max_depth=6,subsample=0.8, colsample_bytree=0.5)
dtrain = xgb.DMatrix(data = as.matrix(train[,-c("Item_Identifier", "Item_Outlet_Sales")]), label= train$Item_Outlet_Sales) dtest = xgb.DMatrix(data = as.matrix(test[,-c("Item_Identifier")]))
```

**Çapraz doğrulama**
Çapraz doğrulama için xgb.cv () işlevini kullanacağız. Bu işlev, xgboost paketinin kendisiyle birlikte gelir. Burada çevrelerin optimal değerini bulmak için çapraz doğrulama kullanıyoruz.

```r
set.seed(112) xgbcv = xgb.cv(params = param_list,  data = dtrain,  nrounds = 1000, nfold = 5, print_every_n = 10, early_stopping_rounds = 30,  maximize = F)
```

Birden çok değerlendirme ölçümü var. Erken durdurmak için test_rmse kullanacaktır. Test_rmse, 30 turda iyileşmediği sürece antrenman yapacak.

### Model Eğitimi
Yukarıdaki ayrıntıya göre, 430. yinelemede en iyi doğrulama / test puanını aldık. Dolayısıyla, XGBoost modelini oluşturmak için nrounds = 430 kullanacağız.

```r
xgb_model = xgb.train(data = dtrain, params = param_list, nrounds = 430)
```

Liderlik puanı: 1154.70

Yine, price_per_unit_wt, Outlet_Years, Item_MRP_Clusters gibi bizim yarattığımız özellikler en önemli değişkenler arasındadır.
